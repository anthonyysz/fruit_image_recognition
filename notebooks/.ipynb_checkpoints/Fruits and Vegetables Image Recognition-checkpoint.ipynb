### 2.1 Imports

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; After importing some necessary packages, I will begin by loading in my training, testing, and validation datasets. Each image will be 256x256 and batch sizes will be 32.

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
from tensorflow.keras import layers
from sklearn.metrics import confusion_matrix
from tensorflow.keras.regularizers import l1
from tensorflow.keras.regularizers import l2
```

```python
#I'm going to import my train, test, and validation datasets in now, using keras.utils.image_dataset_from_directory

height = 256
width = 256
batch_size = 32

testdata = tf.keras.utils.image_dataset_from_directory(
  "../data/raw/test",
    image_size = (height, width),
    batch_size = batch_size,
    shuffle = False)

traindata = tf.keras.utils.image_dataset_from_directory(
  "../data/raw/train",
    image_size = (height, width),
    batch_size = batch_size,
    shuffle = True)

valdata = tf.keras.utils.image_dataset_from_directory(
  "../data/raw/validation",
    image_size = (height, width),
    batch_size = batch_size,
    shuffle = True)

#I had some trouble matching the test images to the validation images later on
#In order to fix that, I had to turn shuffling off for the testing data
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Now that I have my datasets imported, I'd like to make sure that I have the same labels for each. I will print a list of all labels for each dataset, as well as the length of the list of labels.

### 2.2 Data Cleaning

```python
#Here, I'd like to be certain that there are the same labels for each dataset

print(traindata.class_names, len(traindata.class_names))
print(testdata.class_names, len(testdata.class_names))
print(valdata.class_names, len(valdata.class_names))

imagelabels = traindata.class_names

#All looks good! It might be beneficial for me to have these class names saved in the future, 
#so I will save them inside of a variable 'imagelabels'
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; I'd also like to see some of my images from my train data, just to see what I'm working with. I would like to display some labeled images from the training set as an example of the type of pictures I will be dealing with.

## 3. Modeling

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; In the modeling step, I will be creating a model that my data can be trained on in order to predict the label of an image. I will also have my model predict the labels of the testing data, and verify the accuracy of those predictions to see how accurate my model is.

### 3.1 Model Building

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Now that I have imported the data, I can create a model to train the data on. I will be using a keras sequential model. After some trial and error, I created a few functions that will make my code a bit more clean. My compile_model, fit_model, and plot_model_performance functions are recycled for each model I tried (shown below).

```python
#Creating function to compile the model. Uses Adam optimizer at .001 learning rate 
#with sparse categorical cross entropy  as the loss function

def compile_model(model):
    model.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate = .001),
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True),
    metrics = ['accuracy']
    )
```

```python
#Creating function to fit the data to the training data, uses 10 epochs

def fit_model(model):
    return model.fit(traindata, validation_data = valdata, epochs = 10)
```

```python
#Creating function to plot the training and validation accuracy on one graph, with the training and validation loss on another

def plot_model_performance(model):
    train_accuracy = model.history['accuracy']
    val_accuracy = model.history['val_accuracy']
    train_loss = model.history['loss']
    val_loss = model.history['val_loss']
    cycles = range(10)
    
    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.plot(cycles, train_accuracy, label = 'Training Accuracy')
    ax1.plot(cycles, val_accuracy, label = 'Validation Accuracy')
    ax1.legend(loc = 'best')
    ax1.set_ylim(0, 1)
    ax1.set_title('Accuracy')
    
    ax2.plot(cycles, train_loss, label = 'Training Loss')
    ax2.plot(cycles, val_loss, label = 'Validation Loss')
    ax2.legend(loc = 'best')
    ax2.set_ylim(0, 6)
    ax2.set_title('Loss')
    plt.show()
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Initially when I began creating the model, I wanted to have one model that is very consistent and reliable. For image recognition, I decided to go with a series of 2D convolution and 2D max pooling layers, using batch normalization and ReLU as the activation function. For both models, I also included a 2D zero padding of 2 pixels on both height and width. My 2D convolution layers included filters of 64, 32, 16, and 8. Each 2D convolution layer included a kernel size of (3, 3), and each 2D max pooling layer included a pool size of (2, 2). 

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; I decided to create two different models, because I wanted to see how a regularizer would affect the model's accuracy. I feared that I could have been overfitting my model, so I included a model without a regularizer as well as a model with an L2 regularizer. Both are shown below.

```python
#Creating first model with l2 regularizer; calling compile_model, fit_model, and plot_model_performance functions

model1 = tf.keras.Sequential([
    layers.ZeroPadding2D(padding=2),
    layers.Rescaling(1./255.0, input_shape = (height, width, 3)), #rescales inputs from range [0, 255] to [0, 1]
    layers.Conv2D(filters = 64, kernel_size = (3, 3), strides = (1, 1), kernel_regularizer = l2(0.0005)),
    layers.BatchNormalization(axis = -1),
    layers.ReLU(),
    layers.MaxPool2D(pool_size=(2, 2), padding = 'same'),
    layers.Conv2D(filters = 32, kernel_size = (3, 3), strides = (1, 1), kernel_regularizer = l2(0.0005)),
    layers.BatchNormalization(axis = -1),
    layers.ReLU(),
    layers.MaxPool2D(pool_size=(2, 2), padding = 'same'),
    layers.Conv2D(filters = 16, kernel_size = (3, 3), strides = (1, 1), kernel_regularizer = l2(0.0005)),
    layers.BatchNormalization(axis = -1),
    layers.ReLU(),
    layers.MaxPool2D(pool_size=(2, 2), padding = 'same'),
    layers.Conv2D(filters = 8, kernel_size = (3, 3), strides = (1, 1), kernel_regularizer = l2(0.0005)),
    layers.BatchNormalization(axis = -1),
    layers.ReLU(),
    layers.Flatten(),
    layers.Dense(units = len(imagelabels), kernel_regularizer = 'l2')
])

compile_model(model1)
model1fit = fit_model(model1)
plot_model_performance(model1fit)
```

```python
#Creating second model with no regularizer; calling compile_model, fit_model, and plot_model_performance functions

model2 = tf.keras.Sequential([
    layers.ZeroPadding2D(padding=2),
    layers.Rescaling(1./255.0, input_shape = (height, width, 3)), #rescales inputs from range [0, 255] to [0, 1]
    layers.Conv2D(filters = 64, kernel_size = (3, 3), strides = (1, 1)),
    layers.BatchNormalization(axis = -1),
    layers.ReLU(),
    layers.MaxPool2D(pool_size=(2, 2), padding = 'same'),
    layers.Conv2D(filters = 32, kernel_size = (3, 3), strides = (1, 1)),
    layers.BatchNormalization(axis = -1),
    layers.ReLU(),
    layers.MaxPool2D(pool_size=(2, 2), padding = 'same'),
    layers.Conv2D(filters = 16, kernel_size = (3, 3), strides = (1, 1)),
    layers.BatchNormalization(axis = -1),
    layers.ReLU(),
    layers.MaxPool2D(pool_size=(2, 2), padding = 'same'),
    layers.Conv2D(filters = 8, kernel_size = (3, 3), strides = (1, 1)),
    layers.BatchNormalization(axis = -1),
    layers.ReLU(),
    layers.Flatten(),
    layers.Dense(units = len(imagelabels))
])

compile_model(model2)
model2fit = fit_model(model2)
plot_model_performance(model2fit)
```

### 3.2 My Best Model

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; I was able to get excellent performance from both models. In the case of model 1 with the regularizer and model 2 without the regularizer, the validation accuracy was nearly exactly the same after 10 epochs and the training accuracy was within 5%. Where the second model really stood out is in where the loss is next to nothing. In the first model, both the training loss and validation loss are just above .8, while in the second model they are closer to .1 and .25 respectively. The outstandingly low loss value that I was able to produce from my second model will make that my best model for this data, and the model I make my predictions with. Let's save that model below and take a look at the model summary.

```python
#Saving model and printing a summary of each layer

model2.save('../models/best_model', save_format = 'h5')
model2.summary()
```

### 3.3 Predictions

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Now that I have my best model picked out, it is time to make some predictions. First, I stored my models predictions of the test data into a list y_pred. After that, I was able to create a variable called y which I used to get the index of the actual testdata. Finally, I printed my model's accuracy as well as a heatmap of which fruits or vegetables my model guessed incorrectly.

```python
#Getting my predictions and model's accuracy

y_pred = model2.predict(testdata) 
y_pred = np.argmax(y_pred, axis=1)
y = np.concatenate([y for x, y in testdata], axis=0)

acc = accuracy_score(y, y_pred)
print(f'Accuracy on the test set: {100*acc:.2f}%')
```

```python
#Printing a heatmap of where my model incorrectly predicted a label

plt.figure(figsize = (9,7))
conf_mat = confusion_matrix(y, y_pred)
sns.heatmap(conf_mat, annot = True, xticklabels = imagelabels, yticklabels = imagelabels)
plt.xlabel("Predicted Labels")
plt.ylabel("Actual Labels")
plt.show()
plt.show()
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; It seems that my model had some difficulty differentiating between corn and sweet corn. Apart from that, I'd say that it did a serviceable job for this task. Each other fruit or vegetable was guessed correctly at least 7 out of 10 times, with many being guessed correctly a perfect 10 times!

### 3.4 A Deeper look

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; I'd like to take a deeper look into my model and see exactly what it sees when it predicting a label for each image. I would like to look into what my model on an activation layer. Since this will determine how well the model is able to learn, it is worthwile to see what the model sees at this step in the process. I will be taking a look at each of the 16 channels on the 12th layer of my best model. For this instance, I will be inputting the first image of an apple.

```python
#Loading image, printing model's predictions.

img='../data/raw/test/apple/Image_1.jpg'
img = tf.keras.utils.load_img(img, target_size=(height, width))

img_array = tf.keras.utils.img_to_array(img)
img_array = tf.expand_dims(img_array, 0) 
predictions = model2.predict(img_array)
print(predictions)
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Here, we can see that the first item in the array is actually the largest. That means that out predictions is the first item in our list of labels, being the apple. Our model was able to correctly predict the label of this image. Now, let's take a look at the 12th layer: a ReLU hidden layer.

```python
#Printing each channel of 12th layer

layer_outputs = [layer.output for layer in model2.layers]
activation_model = tf.keras.Model(inputs = model2.input, outputs = layer_outputs)
activations = activation_model.predict(img_array)

fig, ax = plt.subplots(4,4, figsize = (9,9))
fig.tight_layout()
j = 0
h = 0

for i in range(16):
    ax[j, h].matshow(activations[12][0, :, :, i], cmap ='plasma')
    ax[j, h].axis("off")
    ax[j, h].set_title("Channel " + str(i), pad = 0)
    h += 1
    if h == 4:
        h -= 4
        j += 1
        
```
